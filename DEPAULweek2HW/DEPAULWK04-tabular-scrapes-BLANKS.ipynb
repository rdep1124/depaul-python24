````python
base_url = "https://apps.health.ny.gov/pubdoh/professionals/doctors/conduct/factions/AlphabetSearchAction.action?alpbhabetSearch=P*&d-49653-p="

base_url
````

````python
urls_lc = [f"{base_url}{url_number}" for url_number in range(1,25)]
urls_lc
````

````python
from random import randrange
import time
````

````python
for url in urls_lc:
    df = pd.read_html(url)
    print(df)
    snoozer = randrange(5,12)
    print(f"Snoozing for {snoozer} seconds before next scrape.")
    time.sleep(snoozer)
````

````python
total_links = len(urls_lc)
````

````python
df_all = []
problem_urls = []
counter = 0

for lastnameswithpurl in urls_lc:
    counter = counter + 1
    print(f"Scraping link {counter} of {total_links}")
    try:
        df = pd.read_html(lastnameswithpurl)
        print(df[0])
        df_all.append(df[0])
    except:
        print(f"Uh oh...problem with {lastnameswithpurl}")
        problem_urls.append(url)
    finally:
        snoozer = randrange(5,12)
        print(f"snoozing for {snoozer} seconds before next scrape")
        time.sleep(snoozer)
print("Done scraping!")
````
